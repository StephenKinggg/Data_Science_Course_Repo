```python
import pandas as pd
df1 = pd.DataFrame({"ID":[1,2,3,5,9],
                   "col1":[1,2,3,4,5],
                  "col2": [6,7,8,9,10],
                    "col3": [11,12,13,14,15],
                    "col4":["apple","orange","banana","strawberry","raspbery"]})
df2 = pd.DataFrame({"ID":[1,1,3,5],
                   "colA":[8,9,10,11],
                  "colB": [12,13,15,17],
                    "col4": ["apple","orange","banana","kiwi"],
                    #"col4":[1,2,3,4]
                  })
```

```python
display(df1, df2)
```

```python
pd.merge(df1, df2)  # ID ler zerinden bakarak birletirdi. 
```

```python
pd.merge(df1,df2, on="ID")
```

```python
pd.merge(df1,df2, on=["ID","col4"])
```

```python
pd.merge(df1,df2, suffixes=["_l","_r"], left_on="col2", right_on= "colA")
```

```python
pd.merge(df1,df2, suffixes=["_l","_r"], left_index=True, right_index=True)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
```
