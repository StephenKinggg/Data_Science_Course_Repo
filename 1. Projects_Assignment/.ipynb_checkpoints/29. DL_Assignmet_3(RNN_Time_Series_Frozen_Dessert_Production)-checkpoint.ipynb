## Preprocessing of Data

```python
len(df)
```

```python
test_size = 24
```

```python
test_ind = len(df)- test_size
test_ind 
```

### Train Test Split

```python
train = df.iloc[:test_ind]
test = df.iloc[test_ind:]
```

```python
train
```

```python
test
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

## Time Series Generator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 12
batch_size = 1
generator = TimeseriesGenerator(train_scaled, train_scaled, length = length, batch_size = batch_size)
```

```python
X, y = generator[0]
```

```python
print(f'Given the Array: \n{X.flatten()}')
print(f'Predict this y: \n {y}')
```

## Modelling & Model Performance

### Import related libraries

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
train_scaled.shape
```

```python
n_features = train_scaled.shape[1]
```

### Creating Model

### with "tanh" activation function

```python
model = Sequential()
model.add(LSTM(100, activation = 'tanh', return_sequences=True, input_shape = (length, n_features)))
model.add(LSTM(50, activation = 'tanh'))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 5)
```

```python
validation_generator = TimeseriesGenerator(test_scaled, test_scaled, length = length, batch_size = batch_size)
```

```python
model.fit_generator(generator,
                    epochs = 20,
                    validation_data = validation_generator,
                    callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

#### Evaluate on Test Data

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:",mse, "\nrmse:",rmse)
```

```python
predictions_scaled = []

first_eval_batch = train_scaled[-length:] # scaled train setinin sonundan aldm.
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):  # 12 tane predict yaptryorum.
    
    # get prediction 1 time stamp ahead
    current_pred = model.predict(current_batch)
    
    # store prediction
    predictions_scaled.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis=1)
```

```python
eval_metrics(test_scaled[:length], predictions_scaled) 
```

#### Inverse Transformations and Comparing

```python
predictions = scaler.inverse_transform(predictions_scaled)
```

```python
compare = test[:12]
```

```python
compare['Predictions'] = predictions
```

```python
compare
```

```python
compare.plot()
```

#### Calculation R2 Score and Error Metrics

```python
eval_metrics(compare.Production, compare.Predictions)
```

## Retrain and Forecasting

Select activation function providing a better score, create your final model with full data, forecast for the next 12 months, and plot this forecast.

```python
full_scaler = MinMaxScaler()
scaled_full_data = full_scaler.fit_transform(df)
```

```python
generator = TimeseriesGenerator(scaled_full_data, scaled_full_data, length = length, batch_size = batch_size)
```

```python
model = Sequential()
model.add(LSTM(100, activation = 'tanh', return_sequences=True, input_shape = (length, n_features)))
model.add(LSTM(50, activation = 'tanh'))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')

model.fit_generator(generator, epochs=15)
```

```python
forecast = []
# Replace periods with whatever forecast length you want

first_eval_batch = scaled_full_data[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)
    
    # store prediction
    forecast.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis = 1)
```

```python
forecast = scaler.inverse_transform(forecast)
```

```python
forecast_index = pd.date_range(start = '2019-09-01', periods = length, freq = 'MS')
```

```python
forecast_index
```

```python
forecast_df = pd.DataFrame(data = forecast, index = forecast_index, columns = ['Forecast'])
```

```python
ax = df.plot()
forecast_df.plot(ax = ax, figsize = (16, 8));
```
