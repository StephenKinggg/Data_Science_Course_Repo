```python
def perfect_number(n):
    m = 10
    result = []
    while True:
        summ = sum([int(i) for i in str(m)])
        if summ == 10:
            result.append(m)
            if len(result) == n:
                break
            else:
                m += 1
        else:
            m += 1
    return result[-1]

perfect_number(2)
```

```python
# combs find all combinations of a list and taken from https://stackoverflow.com/a/54480126
def combs(a):
    if len(a) == 0:
        return [[]]
    cs = []
    for c in combs(a[1:]):
        cs += [c, c + [a[0]]]
    return cs

def split_list(multiset):
    if sum(multiset)%2:
        return []
    else:
        target = sum(multiset)/2
        sums = [items for items in combs(multiset) if sum(items) == target ]
        result=[]
        for items in sums:
            diff = multiset.copy()
            for i in items:
                diff.remove(i)
            result.append((items, diff))
        return result

multiset1 = [15, 5, 20, 10, 35, 15, 10]
print(split_list(multiset1))
```

```python
def pal(text):
    res, resa, son = [], [], []
    for i in range(len(text)):
        res.append(text[i:])
        resa.append(text[i::-1])

    for i in res:
        for j in resa:
            a = 0
            toplam = ''
            if i[a] == j[a]:
                while i[a] == j[a]:
                    toplam += i[a]
                    a += 1
                    if toplam == toplam[::-1]:
                        if toplam not in son:
                            son.append(toplam)
                    if a == len(i) or a == len(j):
                        break
    a = sorted(son, key=lambda x : len(x), reverse = True)
    return a[0]

pal('bananas')
```

```python
# combs find all combinations of a list and taken from https://stackoverflow.com/a/54480126
def combs(a):
    if len(a) == 0:
        return [[]]
    cs = []
    for c in combs(a[1:]):
        cs += [c, c + [a[0]]]
    return cs

def split_list(multiset):
    if sum(multiset)%2:
        return []
    else:
        target = sum(multiset)/2
        sums = [items for items in combs(multiset) if sum(items) == target ]
        result=[]
        for items in sums:
            diff = multiset.copy()
            for i in items:
                diff.remove(i)
            result.append((items, diff))
        return result

multiset1 = [15, 5, 20, 10, 35, 15, 10]
print(split_list(multiset1))
```

```python
from functools import reduce
def least_common(liste):
    a = 2
    res = []
    while True:
        b = False
        for i in range(len(liste)):
            if liste[i] % a == 0:
                b = True
                liste[i] = liste[i] // a
                liste.insert(i, liste[i])
                liste.pop(i + 1)
        if b:res.append(a)
        else:a += 1
        if liste.count(1) == len(liste):break
    result = reduce(lambda x, y: x * y, res)
    return (res, result)

least_common([12, 8, 16])
```

```python
def lcm(num_list):
    
    def multiplier(num):
        if num % i == 0: return int(num / i)
        else: return num

    def least_common_multiple(divisors):
        counter = 1
        for num in divisors: counter *= num; 
        return counter
                        
    divisors = []
    for i in range(2, max(num_list)+1):
        if max(num_list) == 1: break;
        while any([num % i == 0 for num in num_list]):
            num_list = list(map(multiplier, num_list))      
            divisors.append(i)

    return divisors, least_common_multiple(divisors)

print(lcm([12, 8, 16]))
```

```python
def next_letters(text):

    letters_list = list(text)
    new_text = ""

    while letters_list[-1:] == ["Z"]:
      new_text += "A"
      del letters_list[-1]

    if letters_list[-1:] == []:
      new_text += "A"
    else:
      new_text += chr(ord(letters_list[-1]) + 1)
      del letters_list[-1]
      for i in reversed(letters_list):
        new_text += i

    return "".join(reversed(new_text))

next_letters("CZADZZ")
```

```python
def chinese_puzzle(heads, legs):    
    for c in range(heads):
        for r in range(heads):
            if (2 * c) + (4 * r) == legs and c + r == heads:
                chic = c
                rab = r
    return f'Chickens: {chic} || Rabbits: {rab}'

chinese_puzzle(35, 94)
```

```python
def isomorphic_strings(text_1, text_2):
    dictt = {}
    count = 0
    result = True
    for i in text_1:
        if i not in dictt.keys():
            dictt[i] = text_2[count]
            count += 1
        else:
            if dictt[i] == text_2[count]:
                count += 1
                continue
            else:
                result = False
                break
    return result

isomorphic_strings("paper", "title")
```

```python
def is_isomorphic(string1, string2):
  dct = {}
  for i in range(len(string1)):
    if len(string1) != len(string2) :
      return False
      break
    elif string1[i] in dct.keys():
      if dct[string1[i]] == string2[i]:
        continue
      else:
        return False
    else:
      if string2[i] in dct.values():
        return False
      else:
        dct[string1[i]] = string2[i]

  return True

is_isomorphic("eggaa", "addss")
```

```python
def prime(n):
    bol = [i for i in range(1, n + 1) if n % i == 0]
    return True if bol == [1, n] else False

def truncatable(num):
    if '0' in str(num):
        return False
    elif num in [2, 3, 5, 7]:
        return 'Both'
    elif prime(num):
        left = True
        for i in range(1, len(str(num))):
            if not prime(int(str(num)[i:])):
                left = False
                break
        right = True
        for i in range(-2, -len(str(num)) - 1, -1):
            if not prime(int(str(num)[i::-1][::-1])):
                right = False
                break
        if right and left:
            return 'Both'
        elif right:
            return 'Right'
        elif left:
            return 'Left'
        elif not right and not left:
            return False
    else:
        return False

truncatable(9137)
```

sqrt(72)

sqrt(418)

2sqrt(18)

https://edabit.com/challenge/WyttgdGuQGaRBqhhP  Minimum Steps to a Palindrome

Given an incomplete palindrome as a string, return the minimum letters needed to be added on to the end to make the string a palindrome.

Examples
min_palindrome_steps("race")  3
# Add 3 letters: "car" to make "racecar"

min_palindrome_steps("mada")  1
# Add 1 letter: "m" to make "madam"

min_palindrome_steps("mirror")  3
# Add 3 letters: "rim" to make "mirrorrim"
Notes
Trivially, words which are already palindromes should return 0.

```python
def min_palindrome_steps(txt):
    if txt == txt[::-1]:
        return 0
    else:
        a = list(txt)
        a = a + [a[0]]
        count = 1
        if a == a[::-1]:
            return count
        else:
            for i in range(1,len(a)):
                count += 1
                a.insert(-i,a[i])
                if a == a[::-1]:
                    break
    return count
```

```python
def is_harshad(n):
    a = []
    t = n
    while t > 0:
        a.append(t%10)
        t = t//10
    return True if n % sum(a) == 0 else False
```

Your task is to sort a given string. Each word in the string will contain a single number. This number is the position the word should have in the result.
Note: Numbers can be from 1 to 9. So 1 will be the first word (not 0).
If the input string is empty, return an empty string. The words in the input String will only contain valid consecutive numbers.
Examples
"is2 Thi1s T4est 3a"  -->  "Thi1s is2 3a T4est"
"4of Fo1r pe6ople g3ood th5e the2"  -->  "Fo1r the2 g3ood 4of th5e pe6ople"

```python
def order(list1):
    w = list1.split()
    newsen = ""
    i = 1
    while i < len(w) + 1:
        for word in w:
            if str(i) in word:
                newsen += word + ' '
        i += 1
    return newsen.strip()
```
