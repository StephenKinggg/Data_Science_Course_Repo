# Hiyerarik Kmeleme

```python
df = pd.read_csv("USArrests.csv").copy()
df.index = df.iloc[:,0]
df = df.iloc[:,1:5]
#del df.index.name
df.head()
```

```python
df = df.rename_axis(None, axis=0)
```

```python
from scipy.cluster.hierarchy import linkage
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
hc_single = linkage(df, "single")
```

```python
dir(hc_complete)
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(15, 10))
plt.title('Hiyerarik Kmeleme - Dendogram')
plt.xlabel('Indexler')
plt.ylabel('Uzaklk')
dendrogram(
    hc_complete,
    leaf_font_size=10
);  #  ; koymazsak deerler gelir.
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(15, 10))
plt.title('Hiyerarik Kmeleme - Dendogram')
plt.xlabel('Indexler')
plt.ylabel('Uzaklk')
dendrogram(
    hc_complete,
    truncate_mode = "lastp",
    p = 4,  # buradaki say ile oynayarak ka kmeye ayrlacan grebiliriz.
    show_contracted = True
);
```

# PCA

```python
df = pd.read_csv("USArrests.csv").copy()
df.index = df.iloc[:,0]
df = df.iloc[:,1:5]
#del df.index.name
df.head()
```

```python
df = df.rename_axis(None, axis=0)
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
from sklearn.preprocessing import StandardScaler

df = StandardScaler().fit_transform(df) # standartlatrma ilemi yapm olduk.
df[0:5,0:5]
```

```python
# yukardaki df aadaki gibi 3 deiken ile temsil ettik.

from sklearn.decomposition import PCA
pca = PCA(n_components = 3)   # bileen saysn giriyoruz. 2 deiken ihtiyac karladgndan dolay yeterli olduundan dierlerine gerek yok.
pca_fit = pca.fit_transform(df)
```

```python
bilesen_df = pd.DataFrame(data = pca_fit, 
                          columns = ["birinci_bilesen","ikinci_bilesen","ucuncu_bilesen"])
```

```python
bilesen_df.head()
```

```python
pca.explained_variance_ratio_ # nc bileenin aklanan varyans aklama oran dk olduundan sadece bu iki bileen almamz yeterli.
```

```python
pca = PCA().fit(df)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_)) # bu grafik bize bileen saysna karar vermemizde bize fayda salyor.
```
