```python
%%sql
SELECT E.FirstName, SUM(P.Price) Total_Sales 
FROM Orders O 
LEFT JOIN OrderDetails OD ON O.OrderID = OD.OrderID 
LEFT JOIN Products P on OD.ProductID = P.ProductID 
LEFT JOIN Employees E on O.EmployeeID = E.EmployeeID 
GROUP BY O.EmployeeID 
ORDER BY Total_Sales DESC;
```

## 16. Germany'dan verilen siparilerin kategorilerine gre ortalama fiyatlar nelerdir?

```python
%%sql
SELECT * FROM Customers LIMIT 2;
```

```python
%%sql
SELECT * FROM Orders LIMIT 2;
```

```python
%%sql
SELECT * FROM OrderDetails LIMIT 2;
```

```python
%%sql
SELECT * FROM Products LIMIT 2;
```

```python
%%sql
SELECT * FROM Categories LIMIT 2;
```

```python
%%sql
SELECT CategoryName, AVG(P.Price)
FROM (SELECT * FROM Orders O 
       LEFT JOIN Customers C ON O.CustomerID = C.CustomerID 
       WHERE C.Country='Germany' ) K
LEFT JOIN OrderDetails OD ON OD.OrderID = K.OrderID
LEFT JOIN Products P ON P.ProductID=OD.ProductID
LEFT JOIN Categories C ON C.CategoryID = P.CategoryID
GROUP BY CategoryName;
```

## 17. Almanya ya da USA'den verilen siparilerin kategorilerine gre ortalama fiyatlar nelerdir?

```python
%%sql
SELECT CategoryName, AVG(P.Price)
FROM (SELECT * FROM Orders O 
       LEFT JOIN Customers C ON O.CustomerID = C.CustomerID 
       WHERE C.Country='Germany' OR C.Country='USA') K
LEFT JOIN OrderDetails OD ON OD.OrderID = K.OrderID
LEFT JOIN Products P ON P.ProductID=OD.ProductID
LEFT JOIN Categories C ON C.CategoryID = P.CategoryID
GROUP BY CategoryName;
```

## 19. Mterilerin 1997 ylna ait siparilerinin maksimum miktarlarn bulup mterilerin isimleri ile birlikte bykten ke sralaynz.

```python
%%sql
SELECT CustomerName, MAX(Price) maks 
FROM Products P
LEFT JOIN OrderDetails OD ON P.ProductID = OD.ProductID 
LEFT JOIN Orders O on OD.OrderID = O.OrderID 
LEFT JOIN Customers C ON C.CustomerID = O.CustomerID 
WHERE OrderDate like '%1997%' 
GROUP BY CustomerName ORDER BY maks DESC;
```

## 20. Sipari yl 1997 olan siparileri alan alanlar sipari alma saylarna gre sralaynz.

```python
%%sql
SELECT COUNT(*) Order_Sales, E.FirstName 
FROM Orders O
LEFT JOIN Employees E ON O.EmployeeID = E.EmployeeID 
WHERE OrderDate like '%1997%' 
GROUP BY E.FirstName 
ORDER by NO DESC;
```
