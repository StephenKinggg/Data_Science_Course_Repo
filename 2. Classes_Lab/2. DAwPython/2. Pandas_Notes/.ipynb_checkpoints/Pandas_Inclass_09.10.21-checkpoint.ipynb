```python
from sqlalchemy import create_engine  #create_engine ara eleman. python ve database bunun üzerinden konuşacak.
```

```python
temp_db = create_engine("sqlite:///:memory:") # hangi sql bağanacaksak onu yazarız. sqlite gibi. İşlemleri sabit yer vermeeden Ram de tutmak için memory yazıyorum.
```

```python
temp_db.table_names()  # içine baktık tablo yok.
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')  # bu siteden bir tablo indirdik.
```

```python
tables
```

```python
pop = tables[5]  # 5. tabloyu aldık.
```

```python
pop
```

```python
pop.to_sql(name="populations", con=temp_db)  # tabloyu sql atarak nereye bağlayacağımı con ile sağlıyorum.
```

```python
temp_db.table_names()  #geçici database içinde bu tablo var. Bundan sonra normal sql çalışıyor artık.
```

```python
pd.read_sql(sql="populations", con=temp_db)
```

```python
pd.read_sql_query(sql="select Country, Population from populations",con=temp_db)  # içine hangi sütunları çç-ekmek istiyorsak normal sql sorgusu ile yazıp çağırıyoruz.
```
