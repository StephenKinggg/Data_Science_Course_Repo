## PCA with sklearn

## Scaling Data

!!! ÖNEMLİ:
    
PCA da mutlak scale işlemi yapıyoruz ve bu sadece StandardScaler oluyor. Çünkü burada datamızın mutlaka mean değerlerini almamız gerekiyor. Orta hattı çizerek birbirine corr. olmasını engellemek için.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
df_scaled_X = pd.DataFrame(scaled_X, columns = X.columns)
plt.figure(figsize = (23, 7))
sns.heatmap(df_scaled_X.corr().round(2), annot = True)
```

!!! ÖNEMLİ:

Görüldüğü gibi featurelar arasında multicollinearity durum söz konusu. Ancak PCA bu multicollinearity sorunun üstesinden gelebiliyor. 

Eğer bizim datamız linear bir modele uygunsa ve cok güçlü featurelar varsa bu durumda PCA dan çok iyi sonuçlar alabiliriz. Tersi durumda yani data linear değil ve güçlü featurelar yoksa PCA iyi sonuçlar vermez.

Mesela strong feature ve linear ise 100 featuredan 10 component ile %90 gibi bir sonuç alırken, nonlinear bir durumda 100 featuredan 50 component oluştursakda aynı sonucu ulaşamayabiliyoruz. Ancak nonlinear olup bunun yanında strong featurelar varsa bu durumda da PCA iyi sonuçlar verebilir.

Datamızdaki feature ve gözlem sayısına bakar ve min olanı dikkate alarak o kadar component oluşturabiliriz.

## PCA

```python
from sklearn.decomposition import PCA
```

!!! Önemli:

PCA da en önemli parametre olarak n_components kullanacağız. Datamızdaki feature ve raw sayısına bakarak hangisi min. ise onu kullanacağız.

Aşağıda görüldüğü gibi 30 feature 2 componente indirdik.

```python
pca = PCA(n_components=2)
```

```python
principal_components = pca.fit_transform(scaled_X)
```

principal_components  

```python
component_df = pd.DataFrame(data = principal_components, columns = ["first_component", 
                                                                    "second_component"])
component_df
```

## Eigenvalues & Eigenvectors

!!! ÖNEMLİ:

PCA arka planında çalışan öz vektörlerdir. Aşağıda görüldüğü gibi array içinde iki tane liste var. Bunlardan ilki birinci componentin ikincisi ise diğer componenttnidir.

30 featuren bu iki componente katkısı vardır. 30 feature üzerinde datanın % 100 taşır. Genellikle ilk componente taşıdığı bilginin % 40, ikincisine % 25 şeklinde devam eder. Yani birkaç componentte % 100 lük variance % 90 na ulaşmış oluruz.

Featureların üzerlerindeki bilgiyi bu componentlere aktarma işlemi sırasında bu eigenvectors devreye girer. 

Aşağıda First eigenvectors üzerindeki featureların ne kadar katkısı olduğunu görebiliyoruz. Bunu yüzde olarak değil büyüklük olarak düşünelim. 

Her componentte her feature az veya çok mutlaka katkısı vardır.

```python
#Eigenvectors
pca.components_
```

```python
eigenvectors_first_component = pca.components_[0]
eigenvectors_first_component
```

```python
first_eigenvectors = pd.DataFrame(eigenvectors_first_component, index=X.columns, 
                                  columns=["first_eigenvectors"]).sort_values("first_eigenvectors", ascending=False)

first_eigenvectors
```

```python
eigenvectors_second_component = pca.components_[1]
eigenvectors_second_component
```

```python
second_eigenvectors = pd.DataFrame(eigenvectors_second_component, index=X.columns, 
                                   columns=["second_eigenvectors"]).sort_values("second_eigenvectors", ascending=False)

second_eigenvectors
```

```python
pd.concat([first_eigenvectors, second_eigenvectors], axis=1)
```

```python
df_scaled_X.loc[0]  # İlk gözlemi aldım.
```

```python
(df_scaled_X.loc[0] * eigenvectors_first_component).sum() # ilk gözlemi ilk öz vektörler ile çarpıp topladım.
```

```python
(df_scaled_X.loc[0] * eigenvectors_second_component).sum() # ilk gözlemi ikinci öz vektörler ile çarpıp topladım.
```

## K-Means Algorithm

```python
from pyclustertend import hopkins
```

```python
hopkins(component_df, component_df.shape[0])   

# Bu değer ne kadar küçükse o kadar clustere yatkındır.
```

```python
from sklearn.cluster import KMeans

ssd = []

K = range(2,10) # 1 den başlatırsak inertia değeri 1 den 2 ye düşerken çok keskin bir düşüş gösterir ve burada genellikle 2 yi cluster sayısı olarak seçmemiz gerekebilir.

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(component_df)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")
```

Burada bakınca keskin düşüşün durduğu nokta 3 gibi gözüküyor. 
Elbow metodu ise 4 diyor. 
Silhoutte ise 3 diyor.

```python
ssd
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

```python
df_diff.plot(kind='bar')
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(component_df)        # Fit the data to the visualizer
visualizer.show();
```

```python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(component_df)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(component_df, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
sns.scatterplot(x = component_df.first_component, y= component_df.second_component, alpha=0.7)
```

Görsele bakınca buda bana net bir şey söylemedi. 

!!! Önemli:

Ancak elbow da keskin düşüşün durduğu nokta 3 iken, silhoutte 2,3 ve 4 değerleri birbirine çok yakın ise bu durumda 2 de  3 te 4 te alınabilir. 

Biz uzmana bu data 2,3 hatta 4 e yatkın gibi gözüküyor diyerek müşteriden bu konu hakkında destek alırız. 

Bu datada kanser ve değil veya kanserin evreleri şeklinde bir cluster yapısıda söz konusu olabilir.

```python
model = KMeans(n_clusters =2, random_state=42) # müşteriden aldığım bilgiye göre cluster 2 olarak devam ettim.
clusters = model.fit_predict(component_df)
clusters
```

```python
ct = pd.crosstab(y, clusters)
ct
```

## Interpreting PCA results

!!! ÇOK ÖNEMLİ:
    
PCA ların değerlendirilmesi bölümü önemli bir kısımdır. 

Yukarıda bir componentte kanser hastaların sayısının daha fazla olduğunu, diğer componentte ise olmayanları temsil eden featureların sayısının fazla olduğu değerlendirmesi yapmıştık. 

```python
scaled_df = pd.DataFrame(scaled_X, columns=X.columns)
scaled_df.head()
```

```python
combined_df = pd.concat([component_df, scaled_df], axis =1)
correlation = combined_df.corr()
correlation.head()
```

Yukarıdaki tablodan yorumu sadece component ve featurelar üzerinden yapacağımdan dolayı ilk 2 sütun ile diğer satırları istemiyorum.

```python
fig, ax = plt.subplots(figsize=(20, 7))
sns.heatmap(correlation.iloc[:2,2:], cmap='YlGnBu', linewidths=.7, annot=True, fmt='.2f')
plt.show()

# Burada 2 column haricindeki ilk 2 satırın corr aldım.
```

!!! COK ÖNEMLİ:
    
Burada yorum yaparken 0.5 ve üzerindeki corr değerleri dikkate alıncak değerler olacaktır. 

Mesela yarıcap büyürse, alan, çukur, yoğunluk gibi featurelar de artacaktır. Birinci component kanser hastalarının tespitinde yoğun olarak toplandığı bir bileşendir diyebilirim.

İkinci component ise en önemli feature mean fractial dimension ve buna bağlı hatası ve worst fract.dimens. featurelarının kanser hastalarının tespitinde kullanılmayacağını https://pubmed.ncbi.nlm.nih.gov/30705789/ makaleden biliyorum. Bunları dikkate almıyorum. Geriye kalan 0.5 üzeri olanlara dikkat kesiliyorum. Bazı featureların hem hasta hem de olmayanlarda aynı oranda sonuçlar verdiğini görüyorum. 

İkinci componentin ise kanser hastası olmayanların tespitinde kullanılan featureların yoğun bir şekilde toplandığı yani kanser hastalarının tespitinde faydası olmayanların toplandığı bir componenttir diyebiliriz.

Burada değerleri yüksek olanları en önemli featurelar olarak değerlendirmemek gerekir. Sadece 0.5 üzerindekilerin corr. değerlerine bakarak genel bir değerlendirme yapıyoruz.
