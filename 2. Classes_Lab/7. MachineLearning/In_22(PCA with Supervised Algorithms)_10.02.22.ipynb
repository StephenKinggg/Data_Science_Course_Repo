## Split the Data

```python
X = df.drop("Customer_Segment", axis = 1)
```

```python
y = df["Customer_Segment"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y, random_state = 0)
```

## Modeling with Random Forest

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(random_state = 0)
classifier.fit(X_train, y_train)
eval_metric(classifier, X_train, y_train, X_test, y_test)
```

Datamz multiclass olduundan dolay scoring olarak ya micra ya da macro deerlerine bakacaktk. Scorelarm dengeli olduundan dolay micro scorelara baktm. 

Eer bir imbalanced durumu sz konusu olsayd macro, hem imbalanced durumu sz konusu ancak ben says fazla olan classn scorelarn grmek istiyorsam weighted kullanyordum.


```python
from sklearn.model_selection import cross_validate

model = RandomForestClassifier(random_state = 0)

scores = cross_validate(model, X_train, y_train, scoring = ['accuracy', 'precision_micro','recall_micro','f1_micro'], cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

Eer train veya test setinden birinde scorelar tam iken dierinde ciddi fark varsa bu durumda mutlaka CV ile aldmz scorelar kontrol etmemiz gerekir.

Eer btn scorelar dengede ise micro,
bir dengesizlik varsa macro,
bir dengesizlik var ve says fazla olanlar grmek istiyorsam weight kullanyorum.

## Scaling Data

!!! nemli:

PCA da scale kullanmamz gerekir ve bu durumda ise sadece standart scale kullanlr.

CV ve scale yapacamz durumlarda pipeline kullanmak tercih edilir.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X_train_scaled)
x = range(1,14)
plt.plot(x,np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Component count")
plt.ylabel("Variance Ratio");
```

explained_variance_ratio_ grafiini izdik. 

Her bileenin yani componentin zerinde ne kadar bilgi tuttuunu gsteriyor.

```python
pca = PCA(n_components=13).fit(X_train_scaled)
```

```python
my_dict = {"explained_variance":pca.explained_variance_, 
           "explained_variance_ratio":pca.explained_variance_ratio_,
           "cumsum_explained_variance_ratio":pca.explained_variance_ratio_.cumsum()}
```

```python
df_ev = pd.DataFrame(my_dict, index = range(1,14))
df_ev
```

!!! NEML:

Karar verirken %75-95 aral secilir ancak datam linear ve ok gl featurelar varsa bu durumda daha kk deerlerde seilebilir ve bunlar iinde score alnabilir.

n_component iine component says yerine iinde tutmas gereken bilgi miktarn kendimiz girersek min olarak buna salayan component saysn getirir.

pca = PCA(n_components = 0.75) gibi.


Aadaki gibi yapmamn sebebi zvektrler zerinden test seti deerlerimin tm dataya yaylmasn ve data leakage olmasn nlemektir. Scale deki ayn mantk gibi dnebiliriz. 

Sadece train setini verirsek 158 gzlem zerinden eigen vektrdeerleri belirlenirken, tm datay verirsek bu durumda eigen vektr deerleri test setine ait gzlemleride grm olacak. Bunu engellemek iin nce train ve test seti deerlerine ayrp burada train seti deerlerine gre eigen deerlerinin belirlenmesini istiyorum. Ve bylece test seti deerlerini eigen deerlerinin grmesini engellemi oluyoruz. 

pca.components_ deerleri traih zerinden aldm eigen deerleridir.

```python
pca = PCA(n_components = 3) # 0.75 
```

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
df_pca_train = pd.DataFrame(X_train_pca)
df_pca_train = df_pca_train.rename(columns={0:"first_component", 1:"second_component", 2:"third_component"})
df_pca_train.head()
```

```python
df_pca_test=pd.DataFrame(X_test_pca)
df_pca_test = df_pca_test.rename(columns={0:"first_component", 1:"second_component", 2:"third_component"})
df_pca_test.head()
```

## Modelling after PCA

!!! nemli:

max_depths yazmassak scorelarmda ktlemenin ve overfitting olumaya baladn gryorum. Bunu engellemek iin Gridsearch yapmadan nce ilk olarak max_depths deerleri ile oynamak overfitting ile hzl bir ekilde mcadele edebilme imkan salar.

```python
classifier3 = RandomForestClassifier(random_state = 0, max_depth=2)
classifier3.fit(X_train_pca, y_train)
eval_metric(classifier3, X_train_pca, y_train, X_test_pca, y_test)
```

```python
from sklearn.pipeline import Pipeline
```

!!! OK NEML!!!:

Operations ile srasyla nce scale sokup, 

Sonra PCA ile componentleri oluturur, 

Sonra buradan gelenleri RF modeline sokarak score alabiliriz. 

```python
operations = [("scaler", StandardScaler()), ("PCA",PCA(n_components=3)),
              ("RF", RandomForestClassifier(random_state = 0, max_depth=2))]

pipe_model = Pipeline(steps=operations)
                                             
scores = cross_validate(pipe_model, X_train, y_train, scoring = ['accuracy', 'precision_micro','recall_micro','f1_micro'], 
                        cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]                                            
```
