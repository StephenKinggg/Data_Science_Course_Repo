## Modelling

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

Sequential, katmanlardan oluan bir yap oluturmam salar. Dense ise bir layer cinsidir. 

```python
help(Sequential)
```

```python
help(Dense)
```

Her bir dense bir layer temsil ediyor. ki ekilde yaplyor. Ancak ikincisi daha ok tercih ediliyor.

optimizer mincost bularak arlklandrmaya yaryordu.

```python
# First way to create model

model = Sequential([Dense(units = 4, activation = "relu"),
                    Dense(units = 4, activation = "relu"),
                    Dense(units = 4, activation = "relu"),
                    Dense(units = 1)])
```

```python
# Second way to create model (recommended)

model = Sequential()
model.add(Dense(units = 4, activation = "relu"))     # input_dim = X_train.shape[1]
model.add(Dense(units = 4, activation = "relu"))
model.add(Dense(units = 4, activation = "relu"))
model.add(Dense(units = 1))
```

```python
model.compile(optimizer = "rmsprop", loss = "mse")
```

### Choosing an optimizer and loss

Keep in mind what kind of problem you are trying to solve:

    # For a multi-class classification problem
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # For a binary classification problem
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # For a mean squared error regression problem
    model.compile(optimizer='rmsprop',
                  loss='mse')

Burada 22 yi bulmak iin, 700 yani train setindeki gzlem saysn batch_size olan deer 32 ye blyoruz.
Aada loss deerine baknca 250 de ok dtn gryoruz. lk bata ok yksek olmasnn sebebi bata atanan arlklardr. altrdka her seferinde farkl arlklar aldn grebiliriz.

```python
model.fit(x = X_train, y = y_train, batch_size = 32, epochs = 250)
```

## Model Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score
```

yukarda modeli compile ederken metric olarak ne belirtmisem burada o score verir.
model.evaluate arka planda bize predict ilemini otomatik olarak yaptryor.

```python
model.evaluate(X_train, y_train, verbose=0)
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
y_pred = model.predict(X_test)
```

```python
explained_variance_score(y_test, y_pred)
```

```python
my_dict = {"Actual" : y_test, "Pred" : y_pred.reshape(-1,)}
compare = pd.DataFrame(my_dict)
compare.sample(10)
```

```python
sns.scatterplot(x = 'Actual', y = 'Pred', data = compare)
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:", mse, "\nrmse:", rmse)
```

```python
eval_metric(y_test, y_pred)
```

## Final Model and Model Deployment

```python
scaler = MinMaxScaler().fit(X)
```

```python
import pickle
pickle.dump(scaler, open("scaler_fake_reg", 'wb'))
```

```python
X_scaled = scaler.transform(X)
```

```python
final_model = Sequential()

final_model.add(Dense(units = 4, activation = "relu"))     # input_dim = 2 or input_dim = X_train.shape[0]
final_model.add(Dense(units = 4, activation = "relu"))
final_model.add(Dense(units = 4, activation = "relu"))
final_model.add(Dense(1))

final_model.compile(optimizer = "rmsprop", loss = "mse")

final_model.fit(x = X_scaled, y = y, batch_size = 32, epochs = 250)
```

H5 is a file format to store structured data, it's not a model by itself. Keras saves models in this format as it can easily store the weights and model configuration in a single file.

grld gibi yukardaki elde ettiim scorelara ok yakn scorelar elde ettiim iin modelimizi bu ekilde kaydedebilirim. eer olmasayd modeli tekrar train yapmam gerekirdi. h5 uzants ile isim vererek modelimizi kaydettik.

```python
final_model.save('model_fake_reg.h5')  # creates a HDF5 file 'my_model.h5'
```
