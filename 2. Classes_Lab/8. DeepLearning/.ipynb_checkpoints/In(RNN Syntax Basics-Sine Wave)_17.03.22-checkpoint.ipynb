## Preprocessing of Data

### Train Test Split

Burada datann sras ieriinden daha nemli olduundan dolay datay random olarak deil de sral olarak son ksmndan belli bir miktar test datas olarak ayracam.

```python
len(df)
```

```python
test_percent = 0.15
```

```python
len(df) * test_percent
```

```python
test_point = np.round(len(df)*test_percent)
test_point
```

```python
test_ind = int(len(df) - test_point)
test_ind
```

```python
train
```

```python
test
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

## Time Series Generator

Time series generator, CNN de grdmz image generator ile aa yukar ayn eyleri yapyor. 

Amacmz belirlediimiz periyot kadar datay alarak eitip bir tahmin yaptrmak, bu ekilde devam edip datay eitmek. ancak bunun iin generator ile datay bu hale getirmeye alyoruz. 

burada ML deki ve dier DL algoritmalar gibi ayr bir target labelim olmad iin hem data hemde target olarak train_scaled kullanyorum.

length ise burada sequence larmn uzunluu. burada bunlar ksa tuttuk. 

batch size ise bu ekilde hazrladm ka tane datay ayn anda modele sokacam gsterir. genelde tavsiye edilen tek tek yapmak ama byk data setlerinde bu mmkn bir durum deil. 

stride ise kaydrmak demek ve ilk tahminden sonra mesela 2 tane ile 3. tahmin ettikten sonra stride kadar kaydrarak bir sonraki cycle tahmin etmeyi salar.

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
#help(TimeseriesGenerator)
```

nce 
length = 2
batch_size = 1 
stride = 1  aldk ve buna gre deerler aldk. bu durumda generator bana 424 tane data oluturdu nk data setinin son 2 datas ile tahmin ettirecei 3. bir data bulamad iin 424 de brakt. son iki satr kullanamad.
(len(train)-lenght)/(batch_size*stride) formul ile buldu bu sayy.

```python
# define generator
length = 3 # Length of the output sequences (in number of timesteps)
batch_size = 2 # Number of timeseries samples in each batch
stride = 2 # Period between successive output sequences
generator = TimeseriesGenerator(data = train_scaled,
                                targets = train_scaled,
                                length = length,
                                batch_size = batch_size,
                                stride = stride)
```

```python
len(train_scaled)
```

```python
len(generator) # = (len(train)-lenght)/(batch_size*stride) 

# ilk deerlere gre stride 1, son 2 observationi hesaba katmadndan dolay 424 te kald.
# ikinci deerlere gre ise 106 da kald.
```

```python
generator
```

```python
generator[0]  

# grld gibi length 3 ile 3 l bir sequence ile 4. datay ve 
# batchsize 2 ile de bunu 2 li grup olarak yapyor. 
# sonuta bize 2 tane sonu veriyor.
```

```python
# What does the first batch look like?
X, y = generator[0]
```

```python
print(f'Given the Array: \n{X}')
print(f'Predict this y: \n{y}')
```

The more length the more training time

yukardaki sin grafiine baknca tam bir periyot iin 64 satr olmas gerekiyor. yani bir cycle bu kadar satrda tamamlanyor. buna gre length 64 olarak belirledik.

en iyi sonular batch_size 1 iken alrz ancak data bydke bunu artrmak gerekir. burada data uygun olduundan dolay 1 aldk. yani her datada backpropagation yapacak.

```python
length = 64 # Length of the output sequences (in number of timesteps)
batch_size = 1 # Number of timeseries samples in each batch
generator = TimeseriesGenerator(data = train_scaled, targets = train_scaled, length = length, batch_size = batch_size)
```

```python
# What does the first batch look like?
X, y = generator[0]
```

```python
len(generator)  # 426-64/1*1 yapt.
```

```python
print(f'Given the Array: \n{X.flatten()}')
print(f'Predict this y: \n{y}')

# yani buradaki 64 satr modele vererek 1 tane tahmin yapt. many to one diyebiliriz. 
```

## Modelling

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
# We're only using one feature in our time series
n_features = 1
```

```python
seed = 42
```

units 32 alnca model eitim yapmad. 

input_shape aada grld gibi veriyoruz. yani satr ve stn olarak. burada input shape nemli. 64*1 lik bir data geliyor bize. array olarak ince uzun bir data oluturuyor. bir feature tahmin edeceimden dolay output bir olacak.

mesela ANN olsayd burada 200 parametre kullanm olacaktm. 

LSTM iin 1 yerine *4, GRU iin *3 olacakt parametre iin.

```python
# define model
model = Sequential()

# Simple RNN layer. normalde CNN de burada Dense ekliyorduk.
model.add(SimpleRNN(units = 100, activation = "tanh", input_shape=(length, n_features)))

# Final Prediction
model.add(Dense(units = 1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
# params = 1 * ((size_of_input + 1) * size_of_output + size_of_output^2)
1 * ((1+1)*100+(100**2))

# lstm iin 1 yerine 4, GRU iin 3 olacak.
```

```python
model.fit_generator(generator = generator, epochs = 5)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

### Final For Loop to predict step by step 

```python
predictions_scaled = []

first_eval_batch = train_scaled[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    
    # get prediction 1 time stamp ahead
    current_pred = model.predict(current_batch) # sonuc 2 dimensiond.
    
    # store prediction
    predictions_scaled.append(current_pred[0]) # 2 dimensionun ilkini alyorum.
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis = 1)
```

## Early Stopping and Validation Generator

```python
from tensorflow.keras.callbacks import EarlyStopping
```

early stop iin bir validation datas olmas lazm ve bunuda generator iinde hazrlamamz gerekir.

validation generator tanmlarken data ve target olarak test_scaled veriyoruz.

mesela aada length uzunluu 64 olduundan dolay test range en az bundan byk olmalyd. 

```python
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

```python
length = 64 # lenght range have to be shorter than range of test data
            # In this case lenght range = 64, test range = 75
batch_size =1

generator = TimeseriesGenerator(data = train_scaled, targets = train_scaled, length = length, batch_size = batch_size)

validation_generator = TimeseriesGenerator(data = test_scaled, targets = test_scaled, length = length, batch_size = batch_size)
```

```python
len(validation_generator)
```

## LSTM

```python
# define model
model = Sequential()

# LSTM layer
model.add(LSTM(units = 100, activation = "tanh", input_shape = (length, n_features)))

# Final Prediction
model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.summary()
```

```python
# params = 4 * ((size_of_input + 1) * size_of_output + size_of_output^2)
4 * ((1+1)*100+(100**2))
```

LSTM de parametre says 4 kat oluyor. GRU da ise 3 kat oluyor.

```python
model.fit_generator(generator = generator,
                    validation_data = validation_generator,
                    epochs = 20,
                    callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

```python
predictions_scaled = []

first_eval_batch = train_scaled[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    
    # get prediction 1 time stamp ahead 
    current_pred = model.predict(current_batch)
    
    # store prediction
    predictions_scaled.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis = 1)
```

```python
predictions = scaler.inverse_transform(predictions_scaled)
test['LSTM Predictions'] = predictions
test
```

## Retrain and Forecasting

Bu sefer tm data iin tahmin yapacam. scaler tm dataya uyguladm.

```python
full_scaler = MinMaxScaler()
scaled_full_data = full_scaler.fit_transform(df)
```

```python
length = 64 
generator = TimeseriesGenerator(scaled_full_data, scaled_full_data, length = length, batch_size = 1)
```

yukarda 3 epochta modelimin early stop yaptn grdmden dolay burada epoch says 6 olarak aldk.

```python
model = Sequential()
model.add(LSTM(100, input_shape = (length, n_features)))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
model.fit_generator(generator, epochs = 6)
```

```python
forecast = []

first_eval_batch = scaled_full_data[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):
    
    # get prediction 1 time stamp ahead 
    current_pred = model.predict(current_batch)
    
    # store prediction
    forecast.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis = 1)
```

```python
forecast = scaler.inverse_transform(forecast)
```
