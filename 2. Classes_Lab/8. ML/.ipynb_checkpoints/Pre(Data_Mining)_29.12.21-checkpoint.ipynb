```python
from textblob import TextBlob
from sklearn import model_selection, preprocessing, linear_model, naive_bayes, metrics
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn import decomposition, ensemble

import tensorflow as tf

import pandas, xgboost, numpy, textblob, string
from tensorflow.keras.models import Sequential 
from keras.preprocessing import text, sequence
from keras import layers, models, optimizers


from warnings import filterwarnings
filterwarnings('ignore')
```

```python
data.head()
```

```python
data["Sentiment"].replace(0, value = "negatif", inplace = True)
data["Sentiment"].replace(1, value = "negatif", inplace = True)
```

```python
data["Sentiment"].replace(3, value = "pozitif", inplace = True)
data["Sentiment"].replace(4, value = "pozitif", inplace = True)
```

```python
data.head()
```

```python
data = data[(data.Sentiment == "negatif") | (data.Sentiment == "pozitif")]
```

```python
data.groupby("Sentiment").count()
```

```python
df.head()
```

## Metin n leme

```python
#buyuk-kucuk donusumu
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#noktalama iaretleri
df['text'] = df['text'].str.replace('[^\w\s]','')

#saylar
df['text'] = df['text'].str.replace('\d','')

#stopwords
import nltk

#nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in sw))

#seyreklerin silinmesi
sil = pd.Series(' '.join(df['text']).split()).value_counts()[-1000:]
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in sil))

#lemmi
from textblob import Word

#nltk.download('wordnet')
df['text'] = df['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()])) 
```

## Deiken Mhendislii

* Count Vectors
* TF-IDF Vectors (words, characters, n-grams)
* Word Embeddings : kelimenin younluk haritas karlyor.

Tm veri setindeki unique kelimeler karlacak ve bunlarn herbirinin her bir satrda geme sklklar yer alacak.

Herbir satr bir dokman, her bir stunda kelimelerin burada geme sklklarn ifade eder. Buna count vectors denir.

TF(t) = (Bir t teriminin bir dkmanda gzlenme frekans) / (dkmandaki toplam terim says) 

IDF(t) = log_e(Toplam dkman says / iinde t terimi olan belge says)
