```python
#Setler - Klasik Kume Islemleri

# =============================================================================
# difference() ile iki kumenin farkini ya da "-" ifadesi
# intersection() iki kume kesisimi ya da "&" ifadesi
# union() iki kumenin birlesimi
# symmetric_difference() ikisinde de olmayanlari.
# =============================================================================
```

```python
#difference

set1 = set([1,3,5])
set2 = set([1,2,3])

print(set1.difference(set2))

print(set2.difference(set1))

print(set1.symmetric_difference(set2))  # ikisinde de olmayanlar getirir.
```

```python
print(set1 - set2)
print(set2 - set1)
```

```python
#intersection & union

set1 = set([1,3,5])
set2 = set([1,2,3])

print(set1.intersection(set2))
print(set2.intersection(set1))
```

```python
kesisim = set1 & set2  # sonular saklamak istersek bunlar yeni bir deikene atyoruz.
kesisim
```

```python
birlesim = set1.union(set2)
birlesim
```

```python
set1.intersection_update(set2)  # kesiimde gelen elemanlar set1 update yaptk.
set1
```

```python
#Set Sorgu Islemleri

set1 = set([7,8,9])
set2 = set([5,6,7,8,9,10])
```

```python
# iki kumenin kesisiminin bos olup olmadiginin sorgulanmas

set1.isdisjoint(set2)
```

```python
# bir kumenin butun elemanlarinin baska bir kume icerisinde yer alip almadigi

set1.issubset(set2)
```

```python
# bir kumenin bir diger kumeyi kapsayip kapsamadigi

set2.issuperset(set1)
```

```python
# mini uygulama

maaslar = [1000,2000,3000,4000,5000]

def maas_ust(x):   # st maalar iin zaml maa hesab func yazdk.
    print(x*10/100 + x)

def maas_alt(x):   # alt maalar iin zaml maa hesab func yazdk.
    print(x*20/100 + x)

for i in maaslar:
    if i >= 3000:
        maas_ust(i)
    else:
        maas_alt(i)
```

```python
#break & continue : belirli bir art salayan ifadeler yakalandnda dng bitirilmek veya bu durumu grmezden gelmek gerekebilir.
        
# burada maa 3000 e kadar olanlar ile ilgilendiimizi dnelim.
maaslar = [8000,5000,2000,1000,3000, 7000, 1000]

# dir(maaslar)  # maalara uygulanabilecek methodlar grmek istedik.

maaslar.sort()
print(maaslar)

for i in maaslar:
    if i == 3000:
        print("kesildi")  # dngnn kesildiini takip edebilmek iin yazdk.
        break
    print(i)
```

```python
for i in maaslar:
    if i == 3000:
        continue  # bir deer kontrol ifadesine takldnda bu deeri atlamak iin kullanlr. Grld gibi 3000 gelmedi.
    print(i)
```

```python
# Ornek ozellikleri

# Yukardaki hatay gidermek iin yani her bir instance ayr zelliklerden olutuunu python vermemiz gerekir.

class VeriBilimci():  # hem class zelliklerini tanmladk. Hem de aada instance zelliklerini tanmladk.
    bildigi_diller = ["R","PYTHON"]
    bolum = ''
    sql = ''
    deneyim_yili = 0      # self burada rneklemleri temsil eder.
    def __init__(self):  # her bir instance kendi iinde ayr zelliklere sahip olduunu salamak iin
        self.bildigi_diller = []  # instanceler iin tanmlama yapyoruz.
        self.bolum = ''
```

```python
# Ornek Metodlari

# rneklem zerinde alan yani her bir veribilimcinin rendii yeni dili kaydetsin.

class VeriBilimci(): # 
    
    calisanlar = []  # alan isminde bir nesne yani object tanmladm.
    def __init__(self):  # rneklemlerin zelliklerini tanmladk.
        self.bildigi_diller = []
        self.bolum = ''
        
    def dil_ekle(self, yeni_dil) : # dil ekleme iini tek tek yapmak istemediimizden dolay.
        self.bildigi_diller.append(yeni_dil)  # rneklem zelinde ve class altnda alyor.
```

```python
dir(VeriBilimci)
```

```python
ali = VeriBilimci()
print(ali.bildigi_diller)
print(ali.bolum)
```

```python
veli = VeriBilimci()
print(veli.bildigi_diller)
print(veli.bolum)
```

```python
VeriBilimci.dil_ekle
```

```python
VeriBilimci.dil_ekle("R")  # class niteliklerini tanmlarken class ve instance isimlerini ayn ekilde isimlendirmemek gerekir.

# byle bir zellikle class eklemedik. instance zerinden bunu aada deneyelim.
```

```python
ali.dil_ekle("R")
ali.bildigi_diller
```

```python
veli.dil_ekle("Python")
veli.bildigi_diller
```

```python
# Miras Yapilari (inheritance)

# daha nce tanmladmz bir class n zelliklerini yeni class ta kullanmak istiyorsak bunu kullanrz.

class Employees():
    def __init__(self):
        self.FirstName = ""
        self.LastName = ""
        self.Address = ""


class DataScience(Employees):  # burada DataScience() yerine bunlarn rnek zelliklerini kullanmak iin bunun iin o snfn ismini yazyoruz. 
    def __init__(self):
        self.Programming = ""

# veribilimci1 = DataScience()
# veribilimci1. dediimizde hem kendi classnn hem de employees snfnn zellikleri gelir.

class Marketing(Employees):
    def __init__(self):
        self.StoryTelling = ""
```

```python
class Employee_yeni():  # yukardaki ii daha grsel bir ekilde yapmak istersek.
    def __init__(self, FirstName, LastName, Address):
        self.FirstName = FirstName
        self.LastName = LastName
        self.Address = Address


ali = Employee_yeni("a","b","c")
print(ali.FirstName)
print(ali.LastName)
print(ali.Address)
```
