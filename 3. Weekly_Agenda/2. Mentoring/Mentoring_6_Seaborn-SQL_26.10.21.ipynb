```python
# draw graph below
new_tit= titanic.loc[:,["sex","pclass","age","fare","survived"]]

g = sns.PairGrid(new_tit)

g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.scatterplot, linewidths=1, edgecolor="w", s=40)

g = g.add_legend();
```

```python
## 2.yntem
g = sns.PairGrid(titanic, vars=["pclass","age","fare","survived"])

g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.scatterplot, linewidths=1, edgecolor="w", s=40)

g = g.add_legend();
```

```python
# Change default palette to GREEN and RED

g = sns.PairGrid(data=titanic,vars=["pclass","age","fare","survived"], hue="sex", palette={"female":"red","male":"green"})

g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.kdeplot, shade=True)
g = g.map_lower(sns.scatterplot, linewidths=1, edgecolor="w", s=40)

g = g.add_legend();
```

3.1. What is the view? Why we use views?
