```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st 
df = pd.read_csv("data.csv")
```

```python
df.Conversion_A.mean()
```

```python
sns.distplot(df.Conversion_B);
```

```python
t_stat, p_val = st.ttest_ind(df.Conversion_B, df.Conversion_A, alternative="greater")
t_stat , p_val
```

```python
#define sample data
data = [12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29]

#create 95% confidence interval for population mean weight
st.t.interval(alpha=0.9, df=len(data)-1, loc=np.mean(data), scale=st.sem(data))
```

```python
st.t.interval(alpha=0.95, df=len(df.Conversion_A)-1, loc=np.mean(df.Conversion_A), scale=st.sem(df.Conversion_A))
```

```python
st.t.interval(alpha=0.95, df=len(df.Conversion_B)-1, loc=np.mean(df.Conversion_B), scale=st.sem(df.Conversion_B))
```
